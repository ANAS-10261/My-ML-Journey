# Python Basics: Variables

This notebook explains how to use variables in Python.

## 1. What are Variables?

### Theory
Variables are containers for storing data values. Unlike other programming languages, Python has no command for declaring a variable. A variable is created the moment you first assign a value to it.

```python
# Variable assignment
x = 5          # x is of type int
y = "Hello"    # y is of type str
z = 3.14       # z is of type float

print(x)
print(y)
print(z)
```

## 2. Variable Naming Rules

### Theory
A variable can have a short name (like `x` and `y`) or a more descriptive name (`age`, `car_name`).
Rules for Python variables:
- A variable name must start with a letter or the underscore character (`_`).
- A variable name cannot start with a number.
- A variable name can only contain alpha-numeric characters and underscores (`A-z`, `0-9`, and `_`).
- Variable names are case-sensitive (`age`, `Age` and `AGE` are three different variables).

```python
# Legal variable names:
my_variable = "John"
_my_variable = "John"
myVariable = "John"
MYVARIABLE = "John"
myvariable2 = "John"

print("All these are legal variable names.")

# Illegal variable names (uncomment to see the error):
# 2myvar = "John"
# my-var = "John"
# my var = "John"
```

## 3. Getting the Data Type

### Theory
You can get the data type of a variable with the `type()` function.

```python
x = 5
y = "Hello"

print(type(x))
print(type(y))
```

## 4. Case-Sensitivity

### Theory
Variable names are case-sensitive. This means that `a` and `A` are treated as two different variables.

```python
a = 4
A = "Sally"
# A will not overwrite a

print(a)
print(A)
```

## 5. Assign Multiple Values

### Theory
Python allows you to assign values to multiple variables in one line.

```python
# Assign multiple variables at once
x, y, z = "Orange", "Banana", "Cherry"

print(x)
print(y)
print(z)
```
