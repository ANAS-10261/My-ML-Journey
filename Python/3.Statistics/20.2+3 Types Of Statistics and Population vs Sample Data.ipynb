# Types Of Statistics and Population vs Sample Data

This notebook explains the theoretical foundations of two core topics in statistics:

1. Types of statistics
2. Population vs sample data

## 1) Types of Statistics

Statistics is broadly divided into two main types:

- Descriptive Statistics
- Inferential Statistics

### A. Descriptive Statistics

Descriptive statistics deals with methods used to summarize, organize, and present data in a meaningful way. It helps us understand what the data looks like, but it does not make conclusions beyond the given data.

Common descriptive tools include:

- Measures of central tendency: mean, median, mode
- Measures of dispersion: range, variance, standard deviation
- Tabular and graphical methods: frequency tables, bar charts, histograms, box plots

Example idea:
If a class has marks of 50 students, descriptive statistics can summarize average marks, highest and lowest marks, and how spread out the marks are.

### B. Inferential Statistics

Inferential statistics uses sample data to make estimates, decisions, or predictions about a larger population. It is based on probability theory and helps us move from observed data to broader conclusions.

Common inferential tools include:

- Estimation and confidence intervals
- Hypothesis testing
- Regression and correlation analysis
- ANOVA and other model-based methods

Example idea:
A survey of 1,000 voters may be used to estimate the opinion of millions of voters in a country.

### Key Difference: Descriptive vs Inferential

- Descriptive statistics describes data you already have.
- Inferential statistics draws conclusions about data you do not fully observe (the population), using a sample.

Both are essential and are often used together in real analysis workflows.

## 2) Population vs Sample Data

Understanding population and sample is fundamental for inferential statistics.

### A. Population

A population is the complete set of all individuals, objects, or observations that we want to study.

Examples:

- All students in a university
- All voters in a country
- All manufactured bulbs in a factory during one month

A numerical measure that describes a population is called a parameter (for example, population mean).

### B. Sample

A sample is a subset of the population selected for actual observation or measurement.

Examples:

- 500 students selected from the university
- 1,000 voters selected from the country
- 200 bulbs tested from total production

A numerical measure computed from a sample is called a statistic (for example, sample mean).

### Why Use a Sample Instead of the Entire Population?

In many practical situations, studying the full population is difficult because:

- It can be too costly
- It can take too much time
- It may be physically impossible
- Some tests are destructive (for example, stress-testing products)

A well-chosen sample provides a practical and reliable way to learn about the population.

### Population vs Sample: Comparison

- Population: complete group of interest
- Sample: part of that group
- Population measure: parameter
- Sample measure: statistic
- Population study: census
- Sample study: sampling survey

Inferential statistics connects sample statistics to population parameters.

## Good Sampling Principles

To make valid inferences, a sample should be:

- Representative of the population
- Large enough for stable results
- Collected with minimal bias
- Based on clear sampling methods (random, stratified, systematic, etc.)

Poor sampling leads to misleading statistics and incorrect conclusions.

## Summary

- Statistics has two major types: descriptive and inferential.
- Descriptive statistics summarizes observed data.
- Inferential statistics generalizes from sample to population.
- Population is the full group of interest; sample is a subset of it.
- Parameters describe populations, while statistics describe samples.

These ideas form the base for almost every statistical and data science method.
