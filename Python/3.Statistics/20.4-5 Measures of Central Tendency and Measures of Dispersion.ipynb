# Measures of Central Tendency and Measures of Dispersion

This notebook explains two important concepts in statistics:

1. Measures of central tendency
2. Measures of dispersion

Central tendency tells us the typical value of data, while dispersion tells us how spread out the data values are.

## 1) Measures of Central Tendency

A measure of central tendency gives a single value that represents the center of a dataset.

The three most common measures are:

- Mean
- Median
- Mode

### A. Mean (Average)

The mean is the arithmetic average of all observations.

Formula:

$$
\text{Mean} = \frac{\sum x_i}{n}
$$

Where:
- $x_i$ are data values
- $n$ is number of observations

It is sensitive to extreme values (outliers).

### B. Median

The median is the middle value when data is arranged in ascending order.

- If $n$ is odd: median is the middle element
- If $n$ is even: median is the average of the two middle elements

Median is less affected by outliers.

### C. Mode

The mode is the most frequently occurring value in a dataset.

A dataset can be:
- Unimodal (one mode)
- Bimodal (two modes)
- Multimodal (more than two modes)
- No mode (if all values appear equally)

```python
import numpy as np
import pandas as pd

# Example dataset
data = [12, 15, 18, 18, 20, 22, 24, 24, 24, 30]
s = pd.Series(data)

print("Data:", data)
print("Mean:", round(s.mean(), 2))
print("Median:", s.median())
print("Mode:", s.mode().tolist())
```

## 2) Measures of Dispersion

Measures of dispersion show how much data points vary around the center.

Common measures are:

- Range
- Variance
- Standard deviation
- Interquartile range (IQR)
- Coefficient of variation (CV)

### A. Range

Range is the difference between maximum and minimum values.

$$
\text{Range} = \max(x) - \min(x)
$$

It is simple but depends only on two extreme values.

### B. Variance

Variance measures average squared deviation from the mean.

Population variance:
$$
\sigma^2 = \frac{\sum (x_i - \mu)^2}{N}
$$

Sample variance:
$$
s^2 = \frac{\sum (x_i - \bar{x})^2}{n-1}
$$

Larger variance means greater spread in data.

### C. Standard Deviation

Standard deviation is the square root of variance.

$$
\sigma = \sqrt{\sigma^2}, \quad s = \sqrt{s^2}
$$

It is easier to interpret because it has the same unit as the original data.

### D. Interquartile Range (IQR)

IQR measures spread of the middle 50% of data.

$$
\text{IQR} = Q_3 - Q_1
$$

- $Q_1$: 25th percentile
- $Q_3$: 75th percentile

IQR is robust to outliers.

### E. Coefficient of Variation (CV)

CV compares variability relative to mean.

$$
\text{CV} = \frac{\text{Standard Deviation}}{\text{Mean}} \times 100
$$

Useful when comparing variation between datasets with different units or scales.

```python
arr = np.array(data)

data_range = arr.max() - arr.min()
sample_variance = arr.var(ddof=1)
sample_std = arr.std(ddof=1)
q1 = np.percentile(arr, 25)
q3 = np.percentile(arr, 75)
iqr = q3 - q1
cv = (sample_std / arr.mean()) * 100

print("Range:", round(data_range, 2))
print("Sample Variance:", round(sample_variance, 2))
print("Sample Standard Deviation:", round(sample_std, 2))
print("Q1:", round(q1, 2), "Q3:", round(q3, 2), "IQR:", round(iqr, 2))
print("Coefficient of Variation (%):", round(cv, 2))
```

```python
summary = pd.DataFrame({
    "Measure": [
        "Mean", "Median", "Mode",
        "Range", "Sample Variance", "Sample Std Dev",
        "Q1", "Q3", "IQR", "CV (%)"
    ],
    "Value": [
        s.mean(),
        s.median(),
        s.mode().iloc[0],
        data_range,
        sample_variance,
        sample_std,
        q1,
        q3,
        iqr,
        cv
    ]
})

summary.round(2)
```

## Interpretation

- Central tendency gives the typical value of the dataset.
- Dispersion tells how tightly or loosely values are distributed around that center.
- For complete understanding, always report both center and spread.

Example:
- Two classes can have the same mean marks but very different standard deviations.
- The class with lower standard deviation has more consistent performance.

## Summary

- Mean, median, and mode describe central location.
- Range, variance, standard deviation, IQR, and CV describe spread.
- Median and IQR are more robust when outliers are present.
- These measures are foundational for statistical analysis and machine learning.
