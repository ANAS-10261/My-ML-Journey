# Addition Rule (For Mutually and Non-Mutually Exclusive Events)

This notebook explains the addition rule of probability and shows how to use it for mutually exclusive and non-mutually exclusive events.

## What Is the Addition Rule?

The addition rule helps us find the probability that **event A or event B** occurs. In probability, the word **or** means that at least one of the events occurs. It includes the possibility that both events occur.

The general addition rule is:

$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

Where:

- $P(A \cup B)$ is the probability of A or B occurring.
- $P(A)$ is the probability of event A.
- $P(B)$ is the probability of event B.
- $P(A \cap B)$ is the probability that both A and B occur.

## Why Do We Subtract the Intersection?

When A and B can happen together, simply adding $P(A)$ and $P(B)$ counts their shared outcomes twice. We subtract $P(A \cap B)$ once to correct this double counting.

For example, if a student plays both cricket and football, that student belongs to both groups. While finding the probability that a student plays cricket or football, the student should be counted only once.

## Mutually Exclusive Events

Two events are **mutually exclusive** when they cannot occur at the same time. If one event happens, the other event cannot happen.

For mutually exclusive events:

$$P(A \cap B) = 0$$

Therefore, the addition rule becomes:

$$P(A \cup B) = P(A) + P(B)$$

### Examples of Mutually Exclusive Events

- On one roll of a die, getting a 2 and getting a 5.
- On one coin toss, getting heads and getting tails.
- Drawing one card, getting a king and getting a queen.

## Example: Mutually Exclusive Events

A fair die is rolled once. Find the probability of getting a 1 or a 6.

Let:

- A = getting a 1, so $P(A) = \frac{1}{6}$
- B = getting a 6, so $P(B) = \frac{1}{6}$

A die cannot show 1 and 6 on the same roll, so the events are mutually exclusive.

$$P(A \cup B) = P(A) + P(B)$$

$$P(A \cup B) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}$$

So, the probability of getting a 1 or a 6 is $\frac{1}{3}$.

## Non-Mutually Exclusive Events

Two events are **non-mutually exclusive** when they can occur at the same time. These events have one or more outcomes in common.

For non-mutually exclusive events, use the complete addition rule:

$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

### Examples of Non-Mutually Exclusive Events

- Drawing a card that is a heart or a face card.
- Choosing a student who studies mathematics or computer science.
- Selecting a number that is even or greater than 3 from a set of numbers.

In each case, some outcomes may belong to both events.

## Example: Non-Mutually Exclusive Events

A card is drawn from a standard deck of 52 cards. Find the probability of drawing a heart or a face card.

Let:

- A = drawing a heart, so $P(A) = \frac{13}{52}$
- B = drawing a face card, so $P(B) = \frac{12}{52}$
- A face-card heart can be a jack, queen, or king of hearts, so $P(A \cap B) = \frac{3}{52}$

$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

$$P(A \cup B) = \frac{13}{52} + \frac{12}{52} - \frac{3}{52} = \frac{22}{52} = \frac{11}{26}$$

So, the probability of drawing a heart or a face card is $\frac{11}{26}$.

## How to Identify the Correct Rule

Before applying the addition rule, ask one question: **Can both events happen at the same time?**

- If the answer is **no**, the events are mutually exclusive. Add their probabilities directly.
- If the answer is **yes**, the events are non-mutually exclusive. Add their probabilities and subtract the intersection.

This simple check prevents the most common error in addition-rule questions.

## Summary

The addition rule finds the probability that at least one of two events occurs.

- For mutually exclusive events: $P(A \cup B) = P(A) + P(B)$
- For non-mutually exclusive events: $P(A \cup B) = P(A) + P(B) - P(A \cap B)$

Always check whether the events overlap. If they do, subtract the overlap once.
