# Why Sample Variance Is Divided by n-1 and Standard Deviation

This notebook explains two closely related statistics topics:

1. Why sample variance uses `n-1` instead of `n`
2. How standard deviation is related to variance

The ideas are important because sample data is used to estimate the variability of a whole population.

## 1) Variance

Variance measures how far data values are spread out from the mean. It is based on squared deviations, so larger deviations have a stronger effect.

Population variance is written as:

$$
sigma^2 = frac{um (x_i - u)^2}{N}
$$

Where:
- $x_i$ are the data values
- $u$ is the population mean
- $N$ is the population size

When we only have a sample, we usually do not know the true population mean. So we estimate spread using the sample mean.

## 2) Sample Variance

Sample variance is calculated with:

$$
s^2 = rac{um (x_i - ar{x})^2}{n-1}
$$

Where:
- $ar{x}$ is the sample mean
- $n$ is the sample size

The denominator is `n-1`, not `n`, because one degree of freedom is lost after estimating the mean from the same data.

### Why `n-1`?

The sample mean is computed from the sample itself. Once the mean is known, the last deviation is not free to vary independently because all deviations must sum to zero.

So if there are `n` observations, only `n-1` of them can vary freely. This is called **degrees of freedom**.

Using `n` would make the sample variance systematically too small on average. Dividing by `n-1` corrects this bias and gives an unbiased estimate of the population variance.

### Intuition with Deviations

Suppose we have a sample of values and we calculate the mean from them. If most values are above the mean, then the remaining values must compensate by being below the mean.

That constraint reduces the amount of independent information in the sample by one. The `n-1` denominator balances this loss so the variance estimate is not biased downward.

In short:
- `n` is used when the true population mean is known
- `n-1` is used when the mean is estimated from the sample

## 3) Standard Deviation

Standard deviation is the square root of variance. It gives spread in the same unit as the original data, which makes it easier to interpret.

Population standard deviation:

$$
igma = qrt{igma^2}
$$

Sample standard deviation:

$$
s = qrt{s^2}
$$

Because standard deviation is the square root of variance, the `n-1` correction also applies to sample standard deviation through the sample variance.

## 4) Small Example

Consider the sample data:

$$
2,4,6,8
$$

The sample mean is:

$$
ar{x} = rac{2+4+6+8}{4} = 5
$$

Squared deviations from the mean are:

$$
(2-5)^2 = 9,(4-5)^2 = 1,(6-5)^2 = 1,(8-5)^2 = 9
$$

Sum of squared deviations:

$$
9 + 1 + 1 + 9 = 20
$$

Population variance would be:

$$
rac{20}{4} = 5
$$

Sample variance would be:

$$
rac{20}{3} pprox 6.67
$$

Sample standard deviation is:

$$
qrt{6.67} pprox 2.58
$$

```python
import numpy as np

data = np.array([2, 4, 6, 8])

population_variance = data.var(ddof=0)
sample_variance = data.var(ddof=1)
population_std = data.std(ddof=0)
sample_std = data.std(ddof=1)

print("Data:", data.tolist())
print("Population Variance:", round(population_variance, 2))
print("Sample Variance:", round(sample_variance, 2))
print("Population Standard Deviation:", round(population_std, 2))
print("Sample Standard Deviation:", round(sample_std, 2))
```

## 5) Key Takeaways

- Variance measures spread using squared deviations from the mean.
- Sample variance uses `n-1` because one degree of freedom is lost when the mean is estimated from the same sample.
- Dividing by `n-1` gives an unbiased estimate of the population variance.
- Standard deviation is the square root of variance and is easier to interpret because it has the same unit as the data.
- In practice, use population formulas only when the entire population is known; otherwise use sample formulas.
