## Sets


```python
## creation of set
my_set={1,2,3,4}
print(my_set)
##also 
my_empty_set = set()
print(type(my_empty_set))
```

```python
non_duplicate_set = set([1,1,2,2,3,3,3,4,5,6])
print(non_duplicate_set)
```

## Basic Operation

```python
## Adding snd Removing elements
my_set.add(7)
print(my_set)
## Remove
my_set.remove(1)
print(my_set)
```

```python
my_set.add(10)
print(my_set)
my_set.remove(3)
print(my_set)

```

```python
##error found if remove an element not present in the set
my_set.remove(99)
```

```python
## therefore we use the discard function (it find if the element is present and deletes it)
my_set.discard(34)
```

```python
## pop function
removed_elem= my_set.pop()
print(removed_elem)
print(my_set)
```

```python
## clear all the elements in the set with clear()
my_set.clear()
print(my_set)
```

```python
## Set Membership test(Check if element is present in the set or not)
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Union
union_set=set1.union(set2)
print(union_set)

## Intersection
intersection_set=set1.intersection(set2)
print(intersection_set)

## intersection_set => used to copy the intersection back into the set where set.intersection(other_set)
set1.intersection_update(set2)
print(set1)
```

```python
## Difference
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
print(set1.difference(set2))

## similarly
print(set2.difference(set1))
```

```python
## symmetric_difference => used to have the unique elements of both the sets
set1.symmetric_difference(set2)
```

```python
## Sets Methods 
set1={1,2,3,4,5}
set2={3,4,5}
 
# issubset = > to check if first element is the subset of the second set
print(set1.issubset(set2))

print(set1.issuperset(set2))
```

```python
lt=[1,2,2,3,4,4,5]
my_set=set(lt)
print(my_set)
```

```python
## Counting unique words in text

text = " In this tutorial we are discussing about sets"
words=text.split()
print(words)
##convert list of words to get a set of unique words
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
