# Python Basics: Conditional Statements and Loops

This notebook covers:
- `if`, `elif`, `else`
- `for` loop
- `while` loop
- `break`
- `continue`
- `pass`


## 1. Conditional Statements

### Theory
Conditional statements help a program make decisions. The code runs differently based on whether a condition is `True` or `False`.

Python uses `if`, `elif`, and `else`.

```python
# Example of if, elif, and else

age = 18

if age > 18:
    print('You are an adult.')
elif age == 18:
    print('You are exactly 18 years old.')
else:
    print('You are a minor.')
```

## 2. `break` Statement

### Theory
The `break` statement stops a loop immediately, even if more iterations are left.

```python
# Example of break

for i in range(1, 6):
    if i == 4:
        break
    print(i)
```

## 3. `continue` Statement

### Theory
The `continue` statement skips the current iteration and moves to the next one.

```python
# Example of continue

for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

## 4. `pass` Statement

### Theory
The `pass` statement does nothing. It is used when Python needs a statement, but no action is needed yet.

```python
# Example of pass

for i in range(3):
    if i == 1:
        pass
    print(i)
```

## 5. `for` Loop

### Theory
A `for` loop repeats code for each item in a sequence like a list, string, or `range()`.

```python
# Example of for loop

for num in range(1, 6):
    print('Number:', num)
```

## 6. `while` Loop

### Theory
A `while` loop keeps running while its condition is `True`. It is useful when the number of repetitions is not known in advance.

```python
# Example of while loop

count = 1

while count <= 5:
    print('Count:', count)
    count += 1
```

## 7. Combined Example

### Theory
Loops and conditions can be combined to control program flow.

```python
# Combined example using if, for, while, break, continue, and pass

for i in range(1, 6):
    if i == 2:
        continue
    if i == 4:
        break
    print('For loop value:', i)

num = 1
while num <= 3:
    if num == 2:
        pass
    print('While loop value:', num)
    num += 1
```
