# Working with File Paths in Python

## 1. Import Required Libraries
Import the `os` module for traditional path operations and the `Path` class from the `pathlib` module for a modern, object-oriented approach.

```python
import os
from pathlib import Path
```

## 2. The `os.path` Module
Demonstrate basic path operations using the `os` module, such as getting the current working directory with `os.getcwd()`.

```python
current_dir = os.getcwd()
print(f"Current Directory: {current_dir}")
```

## 3. Joining Paths
Show how to correctly join path components using `os.path.join()`. This is essential for creating code that runs on different operating systems (Windows, macOS, Linux).

```python
joined_path = os.path.join(current_dir, "data", "example.txt")
print(f"Joined Path: {joined_path}")
```

## 4. Getting Path Components
Use `os.path.basename()` to get the file name and `os.path.dirname()` to get the directory name from a full path.

```python
file_name = os.path.basename(joined_path)
dir_name = os.path.dirname(joined_path)
print(f"File Name: {file_name}")
print(f"Directory Name: {dir_name}")
```

## 5. Checking Path Existence and Type
Use `os.path.exists()`, `os.path.isfile()`, and `os.path.isdir()` to check if a path exists and whether it's a file or a directory.

```python
print(f"Exists: {os.path.exists(joined_path)}")
print(f"Is File: {os.path.isfile(joined_path)}")
print(f"Is Directory: {os.path.isdir(joined_path)}")
```

## 6. The `pathlib` Module (Modern Approach)
Introduce the `pathlib` module as the modern way to handle filesystem paths. Show how to create a `Path` object and use its methods and attributes to get information.

```python
path_obj = Path(current_dir) / "data" / "example.txt"
print(f"Path Object: {path_obj}")
```

## 7. Navigating the Filesystem with `pathlib`
Demonstrate how to use `pathlib` to get the current working directory (`Path.cwd()`), a user's home directory (`Path.home()`), and how to join paths using the `/` operator.

```python
cwd_path = Path.cwd()
home_path = Path.home()
print(f"CWD: {cwd_path}")
print(f"Home: {home_path}")
```

## 8. Manipulating Paths with `pathlib`
Show how to extract path components like the parent directory (`.parent`), file name (`.name`), stem (`.stem`), and suffix (`.suffix`) from a `Path` object.

```python
print(f"Parent: {path_obj.parent}")
print(f"Name: {path_obj.name}")
print(f"Stem: {path_obj.stem}")
print(f"Suffix: {path_obj.suffix}")
```
