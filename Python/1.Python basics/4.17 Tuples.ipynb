## Tuple Methods


```python
## count(" ") function used to find frequency of the given element
numbers= (1,2,3,4,5,6,1,1)
print(numbers.count(1))
print(numbers.index(2))
```

## Packing and Unpacking tuple

```python
# packing and unpacking a tuple
packed_tuple = 1,"hello",3.14
print(packed_tuple)
```

```python
## Unpacking a tuple
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
##Unpacking a tuple with *
numbers = (1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

## Nested Tuples


```python
## Nested Lists & Nested Tuples
lst=[[1,2,3,4],[6,7,8,9],(1,"Anas",3.14,"a")]
lst[2][1:3]
```

```python
nested_tuple = ((1,2,3),("a","b","c"),(True,False))

print(nested_tuple[0])
print(nested_tuple[1][0:2])
```

```python
# iterating over nested tuples
for subtup in nested_tuple:
    for item in subtup:
        print(item,end=" ")
    print()
```
