# Abstraction in Python

Abstraction is one of the fundamental concepts of Object-Oriented Programming (OOP). Its main goal is to handle complexity by hiding unnecessary internal details of an object from the user, exposing only the essential features.

In Python, abstraction is achieved using **Abstract Base Classes (ABC)** and **abstract methods**. A class that contains one or more abstract methods is considered an abstract class and cannot be instantiated.

## Abstract Base Classes and Methods
Python provides the `abc` module to define abstract base classes. An abstract method is a method that is declared, but contains no implementation in the base class. Child classes are required to inherit and provide implementations for all abstract methods.

```python
from abc import ABC, abstractmethod

# 'Shape' is an abstract base class
class Shape(ABC):
    
    # Abstract method: must be implemented by subclasses
    @abstractmethod
    def area(self):
        pass
    
    # Concrete method: provides a default implementation that can be used or overridden
    def description(self):
        print("I am a generic shape.")


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
        
    # Providing implementation for the abstract method
    def area(self):
        return self.length * self.width


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
        
    # Providing implementation for the abstract method
    def area(self):
        return 3.14 * self.radius ** 2


# Note: We cannot instantiate an abstract class directly
# shape = Shape()  # This would raise a TypeError

# Instantiate subclasses
rect = Rectangle(10, 5)
circle = Circle(7)

rect.description()
print(f"Rectangle area: {rect.area()}")

circle.description()
print(f"Circle area: {circle.area()}")
```
