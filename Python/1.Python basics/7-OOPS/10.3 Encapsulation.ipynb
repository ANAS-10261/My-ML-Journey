# Encapsulation in Python

Encapsulation is one of the fundamental concepts in object-oriented programming (OOP). It describes the idea of bundling data and methods that work on that data within one unit, like a class in Python. This concept is also often used to hide the internal representation, or state, of an object from the outside. This is called information hiding.

## Public Members
Public members (variables and methods) are accessible from outside the class. By default, all members in a Python class are public.

```python
class MyClass:
    def __init__(self, name):
        self.name = name  # public member

    def display(self):
        print(f"Name: {self.name}")

obj = MyClass("John")
obj.display()
print(obj.name)
```

## Private Members
Private members are accessible only within the class. In Python, we denote private members with a double underscore `__` prefix.

```python
class MyClass:
    def __init__(self, name, age):
        self.name = name      # public member
        self.__age = age    # private member

    def display(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.__age}")

obj = MyClass("John", 30)
obj.display()
# This will raise an AttributeError
# print(obj.__age) 
```

## Protected Members
Protected members are accessible within the class and its subclasses. In Python, we denote protected members with a single underscore `_` prefix.

```python
# Base class
class Shape:
    def __init__(self, color):
        self._color = color # protected member

# Derived class
class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def display_color(self):
        print(f"The color of the circle is {self._color}")

c = Circle("Red", 5)
c.display_color()
# Accessing protected member outside of a class is possible but not recommended
print(c._color)
```
