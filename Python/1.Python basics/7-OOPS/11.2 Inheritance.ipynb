# Inheritance in Python

Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a new class (subclass or derived class) to inherit attributes and methods from an existing class (superclass or base class). This promotes code reusability and establishes a relationship between classes.

## Single Inheritance
When a class inherits from a single superclass, it's called single inheritance.

```python
# Base class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

# Derived class
class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

# Derived class
class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

d = Dog("Buddy")
c = Cat("Whiskers")

print(d.speak())
print(c.speak())
```

## Multiple Inheritance
When a class inherits from more than one superclass, it's called multiple inheritance.

```python
# Base class 1
class Father:
    def __init__(self):
        self.father_name = "John"
    def get_father_name(self):
        return self.father_name

# Base class 2
class Mother:
    def __init__(self):
        self.mother_name = "Mary"
    def get_mother_name(self):
        return self.mother_name

# Derived class
class Child(Father, Mother):
    def __init__(self):
        Father.__init__(self)
        Mother.__init__(self)
    def get_parents_names(self):
        return f"Father: {self.get_father_name()}, Mother: {self.get_mother_name()}"

c = Child()
print(c.get_parents_names())
```

## Multilevel Inheritance
When a derived class inherits from another derived class, it's called multilevel inheritance.

```python
# Base class
class Grandfather:
    def __init__(self, grandfather_name):
        self.grandfather_name = grandfather_name

# Intermediate class
class Father(Grandfather):
    def __init__(self, father_name, grandfather_name):
        self.father_name = father_name
        Grandfather.__init__(self, grandfather_name)

# Derived class
class Son(Father):
    def __init__(self, son_name, father_name, grandfather_name):
        self.son_name = son_name
        Father.__init__(self, father_name, grandfather_name)

    def print_family(self):
        print('Grandfather name:', self.grandfather_name)
        print("Father name:", self.father_name)
        print("Son name:", self.son_name)

s1 = Son('Prince', 'Rampal', 'Lal')
s1.print_family()
```

## Hierarchical Inheritance
When more than one derived class are created from a single base, it's called hierarchical inheritance.

```python
# Base class
class Parent:
    def func1(self):
        print("This function is in parent class.")

# Derived class1
class Child1(Parent):
    def func2(self):
        print("This function is in child 1.")

# Derivied class2
class Child2(Parent):
    def func3(self):
        print("This function is in child 2.")

c1 = Child1()
c2 = Child2()
c1.func1()
c1.func2()
c2.func1()
c2.func3()
```

## Hybrid Inheritance
Inheritance consisting of multiple types of inheritance is called hybrid inheritance.

```python
class School:
    def func1(self):
        print("This function is in school.")

class Student1(School):
    def func2(self):
        print("This function is in student 1. ")

class Student2(School):
    def func3(self):
        print("This function is in student 2.")

class Student3(Student1, School):
    def func4(self):
        print("This function is in student 3.")

s3 = Student3()
s3.func1()
s3.func2()
```
