# Control Statements in Python
This notebook covers the basics of control statements in Python.

## 1. `if` statement
The `if` statement is used for conditional execution: a block of code will be executed only if a certain condition is true.

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

## 2. `if-else` statement
The `if-else` statement provides an alternative block of code to be executed if the condition is false.

```python
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

## 3. `if-elif-else` statement
The `if-elif-else` statement allows checking multiple conditions.

```python
x = 5
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")
```

## 4. `for` loop
The `for` loop is used to iterate over a sequence (like a list, tuple, string) or other iterable objects.

```python
# Looping through a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# Using range()
for i in range(5):
    print(i)
```

## 5. `while` loop
The `while` loop executes a block of code as long as a condition is true.

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

## 6. Loop Control Statements

### `break`
The `break` statement is used to exit a loop prematurely.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

### `continue`
The `continue` statement is used to skip the rest of the code inside a loop for the current iteration and move to the next one.

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

### `pass`
The `pass` statement is a null operation; nothing happens when it executes. It is used as a placeholder for future code.

```python
def my_function():
    pass  # Placeholder for future implementation

for i in range(5):
    if i == 3:
        pass
    print(i)
```
