**Iterators**

An iterator is an object that contains a countable number of values. It is an object that can be iterated upon, meaning that you can traverse through all the values. In Python, an iterator is an object which implements the iterator protocol, which consist of the methods `__iter__()` and `__next__()`.

```python
## iter() and next()

my_list = [1, 2, 3, 4]
iterator = iter(my_list)

print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

```python
## Iterating using a loop

my_tuple = ("apple", "banana", "cherry")
for x in my_tuple:
    print(x)
```

```python
## Iterating through a string

my_string = "Hello"
str_iterator = iter(my_string)

print(next(str_iterator))
print(next(str_iterator))
```

```python
## Custom Iterator

class MyNumbers:
    def __iter__(self):
        self.a = 1
        return self

    def __next__(self):
        if self.a <= 5:
            x = self.a
            self.a += 1
            return x
        else:
            raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
    print(x)
```
