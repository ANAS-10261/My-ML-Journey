#Generators 

Generators are a simpler way to create iterators. They use the `yield` keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.

```python
def square_using_return(n):
    for i in range(3):
        return i**2
```

```python
square_using_return(3)
```

```python
## Implementing generators
def square_using_yield(n):
    for i in range(3):
        yield i**2
```

```python
square_using_yield(3)
```

```python
for i in square_using_yield(3):
    print(i)
```

```python
a=square_using_yield(3)
a
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```
