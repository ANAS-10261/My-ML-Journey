# Decorators

Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or
class method. They are commonly used to add functionality to functions or methods without modifying their
actual code. This lesson covers the basics of decorators, including how to create and use them.

# Function Copy

```python
## function Copy

def welcome():
    return "Welcome Anas :)"

welcome()
```

```python
## Copying the function welcome to wel
wel = welcome
wel()  ## Created the copy 

del welcome  ##deleted the original function
wel()   ## After deletion the duplicate function is still remained

```

# Closures

```python
## Closures 

def main_welcome(msg):
    ## msg ="welcome"
    ## Sub method can access this variable
    def sub_welcome_method():  ## Sub method or method within the main_welome 
        print("Welcome Anas :)")
        print(msg)
        print("Please Learn Consistently")
        

    return sub_welcome_method()    
```

```python
main_welcome("Welcome Everyone")
```

```python
## We Can also use functions using func()

def main_welcome_function(func):
    
    def sub_welcome_function():
        print("Welcome Anas :)")
        func("Lets Finish this course")  ## basically func geting replaced with print
        print("Please Learn Consistently")
    return sub_welcome_function()
```

```python
main_welcome_function(print)
```

```python
def main_welcome_function1(func,lst):
    
    def sub_welcome_function():
        print("Welcome Anas :)")
        print(func(lst))
        print("Please Learn Consistently")
    return sub_welcome_function()
```

```python
main_welcome_function1(len,[1,2,3,4,5])
```

# Decorators

```python
## Using Decorators we can use 
@main_function  ##Using this decorator the function is entered inside the parent function
def course_introduction():
    print("This is not a basic course")
```

```python
## Decorators 

def my_decorator(func):
    def wrapper():
        print("before the function...")
    
        func()
    
        print("After the function...")
    return wrapper()
```

```python
@my_decorator
def say_hello():
    print("Heyy!!")
```

```python
## Decorators with arguments

def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for i in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello")
```

```python
say_hello()
```
