**Import Modules and Packages**

Used for organizing and reuse of code 

```python
import math
print(math.sqrt(25))
```

```python
from math import sqrt,pi
print(pi*4*4)
print(sqrt(25))
```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
from math import *
print(sqrt(25))
print(pi)
```

Making Custom Packages

```python
from package.maths import *
print(addition(2,3))
print(substraction(4,3))
```

```python
from package import maths
print(maths.addition(2,3))
print(maths.substraction(10,5))
```

Make Subpackages

```python
from package.maths import *
from package.subpackages.multiply import multiply
print(addition(2,3))
print(substraction(4,3))
print(multiply(2,3))
```
