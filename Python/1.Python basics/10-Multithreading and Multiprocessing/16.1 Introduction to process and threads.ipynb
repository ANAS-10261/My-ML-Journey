# Introduction to Processes and Threads

This notebook introduces the two main ways Python can run work concurrently: processes and threads. It covers the core ideas, the differences between them, and simple examples you can study or run locally.

## What are Processes and Threads?

A **process** is an independent running program with its own memory space.

A **thread** is a smaller execution unit inside a process. Multiple threads can share the same memory space within that process.

Python provides the `multiprocessing` module for processes and the `threading` module for threads.

## Core Concepts

- **Process**: Has its own memory and Python interpreter state.
- **Thread**: Shares memory with other threads in the same process.
- **Concurrency**: Managing multiple tasks that make progress over the same time period.
- **Parallelism**: Running tasks at the same time on different CPU cores.
- **GIL**: In standard CPython, the Global Interpreter Lock limits true parallel execution of Python bytecode in threads.

## Process vs Thread

| Feature | Process | Thread |
| --- | --- | --- |
| Memory | Separate memory space | Shared memory within the process |
| Overhead | Higher | Lower |
| Communication | Slower, requires IPC | Faster, easier sharing |
| Best for | CPU-heavy work | I/O-heavy work |
| Failure impact | One process usually does not crash others | One bad thread can affect the whole process |

## How the Python `threading` Module Works

Threads are useful when your program spends time waiting for external work such as file operations, network calls, or database requests.

Because threads share memory, they can coordinate through shared variables, but that also means you must be careful about race conditions and use synchronization tools such as locks when needed.

```python
# Example: creating and running threads
import threading
import time

def task(name, delay):
    print(f'Starting {name}')
    time.sleep(delay)
    print(f'Finished {name}')

threads = [
    threading.Thread(target=task, args=('Thread-1', 1)),
    threading.Thread(target=task, args=('Thread-2', 1)),
]

for thread in threads:
    thread.start()

for thread in threads:
    thread.join()

print('All threads completed')
```

## How the Python `multiprocessing` Module Works

Processes are useful when you want true parallel execution for CPU-bound work. Each process has its own memory space, so work is isolated and can use multiple CPU cores.

On Windows, multiprocessing code should be protected with the `if __name__ == '__main__':` guard.

```python
# Example: creating and running processes
from multiprocessing import Process, current_process

def worker():
    print(f'Running in {current_process().name}')

if __name__ == '__main__':
    process = Process(target=worker)
    process.start()
    process.join()
    print('All processes completed')
```

## When to Use Each One

- Use **threads** for I/O-bound tasks such as downloads, requests, or waiting on external services.
- Use **processes** for CPU-bound tasks such as number crunching, image processing, or heavy data transformation.
- Use **synchronous code** when the task is simple and concurrency is not needed.

## Important Notes

- Threads share memory, so shared data needs careful coordination.
- Processes are heavier to start but avoid many GIL limitations for CPU work.
- Always keep Windows multiprocessing code inside a main guard.
