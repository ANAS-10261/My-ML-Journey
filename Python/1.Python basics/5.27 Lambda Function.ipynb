## Lambda Fucntions

```python
#Syntax
lambda arguments: expression
```

```python
def addition(a,b):
    return a+b
addition(2,3)
```

```python
additon= lambda a,b: a+b
print(type(additon))
print(additon(5,6))
```

```python
def even(num):
    if num%2==0:
        return True
    
even(24)
```

```python
even1=lambda num: num%2==0
print(even1(10))
```

```python
def addition(x,y,z):
    return x+y+z
addition(10,20,30)
```

```python
addition1=lambda x,y,z: x+y+z
addition1(10,20,30)
```

```python
## map() function - applies a function to all items in a list
numbers = [1,2,3,4,5,6]
def square(number):
    return number**2
square(2)
```

```python
list(map(lambda x: x**2 , numbers))
```
