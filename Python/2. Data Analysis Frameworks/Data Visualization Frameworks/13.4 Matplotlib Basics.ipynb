```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. Basic Line Plot
A line plot is the most basic and common type of chart. It's often used to show trends over continuous data, like time.

```python
# Create some simple data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Plot the data
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis (Time/Categories)")
plt.ylabel("Y-axis (Values)")

# Show the plot
plt.show()
```

## 2. Scatter Plot
Scatter plots are used to observe relationships between variables. They use dots to represent values for two different numeric variables.

```python
# Generate random data
x_scatter = np.random.rand(50) * 100
y_scatter = np.random.rand(50) * 100

# Create a scatter plot
plt.scatter(x_scatter, y_scatter, color='r', alpha=0.5)

plt.title("Scatter Plot Example")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")

plt.show()
```

## 3. Bar Chart
Bar charts represent categorical data with rectangular bars. They are perfect for comparing different groups.

```python
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
quantities = [25, 40, 15, 30]

# Create a bar chart
plt.bar(categories, quantities, color=['red', 'yellow', 'darkred', 'brown'])

plt.title("Fruit Inventory")
plt.xlabel("Fruit Type")
plt.ylabel("Quantity")

plt.show()
```

## 4. Histogram
Histograms are used to visualize the distribution of a continuous dataset. It divides the data into "bins" and shows the frequency of data points in each bin.

```python
# Generate a normal distribution of data
data = np.random.randn(1000)

# Create a histogram with 20 bins
plt.hist(data, bins=20, color='green', edgecolor='black')

plt.title("Normal Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

## 5. Subplots
Subplots allow you to place multiple charts side-by-side or stacked in a single figure.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4)) # 1 row, 2 columns

# First Plot
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title("Line Plot")

# Second Plot
ax2.bar(['A', 'B', 'C'], [3, 1, 4], color='orange')
ax2.set_title("Bar Chart")

plt.tight_layout() # Adjusts layout so labels don't overlap
plt.show()
```
