# Data Visualization in Seaborn

Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Loading a Sample Dataset

```python
# Load the 'tips' dataset
tips = sns.load_dataset('tips')
tips.head()
```

## Scatter Plot (Relational Plot)

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='smoker')
plt.title('Total Bill vs Tip')
plt.show()
```

## Distribution Plot (Histogram)

```python
sns.histplot(data=tips, x='total_bill', kde=True, hue='time')
plt.title('Distribution of Total Bill')
plt.show()
```

## Categorical Plot (Box Plot)

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='day', palette='Set2')
plt.title('Total Bill by Day')
plt.show()
```

## Violin Plot
A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared.

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, palette='muted')
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

## Pair Plot
A pair plot allows us to see both distribution of single variables and relationships between two variables. It creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame.

```python
sns.pairplot(tips, hue='sex', palette='husl')
plt.show()
```

## Matrix Plot (Heatmap)
Heatmaps display data in a 2D matrix format, using color to represent the values. They are excellent for visualizing correlation matrices.

```python
# Calculate correlation only for numeric columns
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_tips.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```
