# More on Data Manipulation in Pandas

Building upon the basics, we will now explore more advanced data manipulation techniques such as filtering, handling missing values, grouping, and applying functions.

```python
import pandas as pd
import numpy as np
```

## 1. Creating a Sample DataFrame

```python
data = {
    'Name': ['Anas', 'John', 'Jack', 'Sam', 'Hamdan'],
    'Age': [19, 20, 25, 30, 15],
    'Salary': [50000, 40000, 35000, 60000, 20000],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR']
}
df = pd.DataFrame(data)
df
```

## 2. Filtering Data (Boolean Indexing)

```python
# Filter people older than 20
df_filtered = df[df['Age'] > 20]
df_filtered
```

```python
# Filter with multiple conditions (Age > 18 and Department is IT)
df_multi = df[(df['Age'] > 18) & (df['Department'] == 'IT')]
df_multi
```

## 3. Sorting Data

```python
# Sort by Age in descending order
df_sorted = df.sort_values(by='Age', ascending=False)
df_sorted
```

## 4. Handling Missing Data (NaN values)

```python
# Adding some missing values for demonstration
df_missing = df.copy()
df_missing.loc[1, 'Salary'] = np.nan
df_missing.loc[3, 'Age'] = np.nan
df_missing
```

```python
# Check for missing values
df_missing.isna().sum()
```

```python
# Drop rows with missing values permanently
df_dropped = df_missing.dropna()
df_dropped
```

```python
# Fill missing values with a specific value (e.g., mean/median)
df_filled = df_missing.copy()
df_filled['Salary'] = df_filled['Salary'].fillna(df_filled['Salary'].mean())
df_filled['Age'] = df_filled['Age'].fillna(df_filled['Age'].median())
df_filled
```

## 5. Grouping and Aggregation

```python
# Group by Department and calculate average properties
grouped_df = df.groupby('Department').mean(numeric_only=True).reset_index()
grouped_df
```

```python
# Applying multiple aggregate functions
agg_df = df.groupby('Department').agg({
    'Salary': ['mean', 'sum'],
    'Age': 'max'
})
agg_df
```

## 6. Applying Functions (`apply`)

```python
# Apply a lambda function to increase everyone's salary by 10%
df['Bonus Salary'] = df['Salary'].apply(lambda x: x * 1.10)
df
```

## 7. Concatenation

```python
data2 = {
    'Name': ['Emily', 'Michael'],
    'Age': [28, 35],
    'Salary': [55000, 70000],
    'Department': ['IT', 'Finance']
}
df2 = pd.DataFrame(data2)

# Append new rows to the existing DataFrame
df_concat = pd.concat([df, df2], ignore_index=True)
df_concat
```
