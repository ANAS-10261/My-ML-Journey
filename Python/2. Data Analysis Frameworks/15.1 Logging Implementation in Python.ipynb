# Logging Implementation in Python

Python's built-in `logging` module lets you record messages, track events, and debug applications without relying on `print()` statements.

This notebook shows how to configure logging, write messages at different levels, save logs to a file, and capture exceptions.

## 1. Basic Logging Configuration

Start with a simple configuration that sends log messages to the console.

```python
import logging

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
)

logging.debug('This is a debug message and will not appear.')
logging.info('Application started successfully.')
logging.warning('This is a warning message.')
logging.error('This is an error message.')
```

## 2. Logging Levels

The most common levels are `DEBUG`, `INFO`, `WARNING`, `ERROR`, and `CRITICAL`.

```python
levels = [
    ('DEBUG', logging.DEBUG),
    ('INFO', logging.INFO),
    ('WARNING', logging.WARNING),
    ('ERROR', logging.ERROR),
    ('CRITICAL', logging.CRITICAL),
]

for level_name, level_value in levels:
    logging.log(level_value, f'Example message for {level_name} level')
```

## 3. Log Messages to a File

You can also save logs to a file for later review.

```python
file_logger = logging.getLogger('file_logger')
file_logger.setLevel(logging.INFO)

file_handler = logging.FileHandler('app.log', mode='w')
file_handler.setFormatter(logging.Formatter('%(asctime)s - %(levelname)s - %(message)s'))

if not file_logger.handlers:
    file_logger.addHandler(file_handler)

file_logger.info('This message is written to the log file.')
file_logger.warning('This warning is also written to the log file.')
print('Log messages have been saved to app.log')
```

## 4. Log Exceptions with `exception()`

Use `logging.exception()` inside an `except` block to record the error message and traceback.

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    logging.exception('An exception occurred while dividing numbers.')
```

## 5. Final Notes

Logging is preferred over `print()` in real applications because it supports severity levels, structured output, and log destinations such as files, streams, and external monitoring systems.
