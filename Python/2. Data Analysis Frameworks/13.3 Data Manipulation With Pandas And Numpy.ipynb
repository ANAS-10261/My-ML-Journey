# Data Aggregating And Grouping

```python
df = df_csv.head(5)
df
```

```python
df['Category'] = ['Group1','Group3','Group1','Group2','Group3']
```

```python
df['SALES'] = df['SALES'].fillna(df['SALES'].mean())
```

```python
## Data Grouping
grouped_mean = df.groupby('Category')['SALES'].mean()
print(grouped_mean)
```

```python
## According to multiple categorical values 
category_col = df.pop('Category')  ## Moving the Category Column to the front index = 0
df.insert(0, 'Category', category_col)
category_col = df.pop('SALES')
df.insert(1,'SALES',category_col)
df.loc[[1,2,4],'PRODUCTLINE'] = 'CAR'
df
```

```python
grouped_sum = df.groupby(['Category','PRODUCTLINE'])['SALES'].sum()
print(grouped_sum)
```

```python
grouped_mean = df.groupby(['Category','PRODUCTLINE'])['SALES'].mean()
print(grouped_mean)
```

```python
## Aggregate Multiple Functions 
grouped_agg = df.groupby('PRODUCTLINE')['SALES'].agg(['mean','sum','count']) 
print(grouped_agg)
```

# Merging And Joining Data Frames

```python
# Create Sample DATA FRAMES
df1 = pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'], 'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merging on the basis of the columns('Key' column)

pd.merge(df1,df2,on='Key',how='inner')  ## inner join causes same element from key column for merge
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```

```python
pd.merge(df1,df2,on="Key",how="left")
```
