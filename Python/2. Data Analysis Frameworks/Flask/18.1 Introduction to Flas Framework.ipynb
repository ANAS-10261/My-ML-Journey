# Introduction to Flask Framework

This notebook introduces the Flask microframework, covering core concepts, architecture, and a minimal example you can run locally.

## What is Flask?

Flask is a lightweight WSGI web application microframework for Python. It's designed to be simple, extensible, and easy to get started with. Flask provides routing, request/response handling, templating via Jinja2, and an ecosystem of extensions for features like database integration, authentication, and migrations.

Key design goals: minimal core, explicit is better than implicit, and freedom to choose components (ORM, form library, etc.).

## Core Concepts

- **Application**: The Flask app object (`Flask(__name__)`) is the central registry for routes, configuration, and extensions.
- **Routes / Views**: Functions decorated with `@app.route()` handle specific URL paths and return responses.
- **Request / Response**: Flask uses Werkzeug to represent incoming requests and responses; you access request data via the `request` object.
- **Templates**: Jinja2 templates render HTML with placeholders and control structures.
- **WSGI**: Flask apps are WSGI applications and can run on any WSGI-compliant server (e.g., Gunicorn, uWSGI).
- **Extensions**: Flask has many extensions (Flask-SQLAlchemy, Flask-Migrate, Flask-Login, etc.) for common tasks.

## Architecture & Request–Response Cycle

1. A client sends an HTTP request to the server (e.g., `GET /`).
2. The WSGI server passes the request to the Flask application (a WSGI callable).
3. Flask matches the request URL to a registered route and calls the associated view function.
4. The view returns a response object or string; Flask converts it to a proper HTTP response.
5. The server sends the HTTP response back to the client.

Middlewares, error handlers, and before/after request hooks can run at different stages of this cycle.

```python
# Minimal Flask application (save this as app.py to run outside the notebook)
from flask import Flask, render_template_string, request

app = Flask(__name__)

@app.route('/')
def index():
    user = request.args.get('user','World')
    return render_template_string('<h1>Hello, {{ name }}!</h1>', name=user)

# if __name__ == '__main__':
#     app.run(debug=True, port=5000)

# Note: Running a Flask app inside a notebook is not recommended for production. Save and run with the flask CLI or a Python interpreter.
```

## Running the app (local development)

Save the example above to `app.py`, then in a terminal run:

- Windows PowerShell:
  `$env:FLASK_APP="app.py"`
  `flask run --port 5000`

- Windows Command Prompt (cmd):
  `set FLASK_APP=app.py`
  `flask run --port 5000`

Open `http://127.0.0.1:5000/?user=YourName` in your browser to see the app.

## When to use Flask

- Small to medium APIs and web apps where flexibility and simplicity matter.
- Prototyping services quickly and iterating fast.
- Microservices or apps where you want to pick specific components rather than an all-in-one framework.

For larger, batteries-included projects with heavy built-in conventions, consider Django. For lightweight, composable apps, Flask is an excellent choice.

## Security Notes

- Do not enable `debug=True` in production.
- Validate and sanitize user input; use secure session management and HTTPS in production.
- Use vetted Flask extensions for authentication and database access.
