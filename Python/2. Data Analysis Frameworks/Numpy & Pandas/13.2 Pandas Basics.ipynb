# Pandas

Pandas is a fundamental library for data manipulation and analysis in Python. It provides high-performance, easy-to-use data structures like Series and DataFrames. A Series is a one-dimensional array-like object,
while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure
with labeled axes (rows and columns).

# Series

```python
'''
Series 
A panda series is like a 1 Dimensional Array like object that can hold any datatype
similar to a column in a table'''

import pandas as pd
data=[1,2,3,4,5]

series = pd.Series(data)
print("Series \n", series)
```

```python
data_dict = { 'a':1 , 'b':2 , 'c':3}
series_dict =pd.Series(data_dict)

print(series_dict)
```

```python
data = [10,20,30]
index = ['a','b','c']

pd.Series(data,index=index)
```

# Data Frames

```python
## DataFrames
## Creating a Dataframe from a dictionary of lists

data ={
    'Name': ['Anas','John','Jack'],
    'Age':[19,20,25],
    'City':['Nagpur','Los Angeles','New York']
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## Creating a Data Frame from a list of Dictionaries

data1 =[
    {'Name':'Anas' , 'Age' : 19 , 'City': 'Dubai'},
    {'Name':'Hamdan' , 'Age' : 15 , 'City': 'Sharjah'},
    {'Name':'Sam' , 'Age' : 30 , 'City': 'New York'},
    {'Name':'Samadhnya' , 'Age' : 25 , 'City': 'Los Angeles'},  
]
df = pd.DataFrame(data1)
print(df)
print(type(df))
```
