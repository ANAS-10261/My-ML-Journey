# CRUD Operations with SQLite3 and Python

SQLite is a C library that provides a lightweight disk-based database that doesn't require a separate server process. This notebook covers the basic CRUD (Create, Read, Update, Delete) operations using the built-in `sqlite3` module in Python.

```python
import sqlite3

# 1. Connect to SQLite (creates a new file named 'example.db' if it doesn't exist)
conn = sqlite3.connect('example.db')

# Create a cursor object using the cursor() method
cursor = conn.cursor()

# 2. CREATE TABLE
cursor.execute('''
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        age INTEGER,
        email TEXT UNIQUE NOT NULL
    )
''')
conn.commit()

print("Database and table created successfully.")
```

## 1. Create (Insert Data)
Let's insert some data into the `users` table.

```python
# Single Insert
cursor.execute("INSERT INTO users (name, age, email) VALUES ('Alice', 28, 'alice@example.com')")

# Multiple Insert
users_data = [
    ('Bob', 22, 'bob@example.com'),
    ('Charlie', 35, 'charlie@example.com'),
    ('Diana', 30, 'diana@example.com')
]
cursor.executemany("INSERT INTO users (name, age, email) VALUES (?, ?, ?)", users_data)
conn.commit()

print("Records inserted successfully.")
```

## 2. Read (Query Data)
Retrieve the data using `SELECT` queries.

```python
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()

print("ID | Name | Age | Email")
print("-" * 30)
for row in rows:
    print(f"{row[0]}  | {row[1]} | {row[2]}  | {row[3]}")
```

## 3. Update (Modify Data)
Use the `UPDATE` statement to modify existing records.

```python
# Update Alice's age
cursor.execute("UPDATE users SET age = 29 WHERE name = 'Alice'")
conn.commit()

# Verify the update
cursor.execute("SELECT * FROM users WHERE name = 'Alice'")
print("Updated Record:", cursor.fetchone())
```

## 4. Delete (Remove Data)
Use the `DELETE` statement to remove records from the table.

```python
# Delete Bob's record
cursor.execute("DELETE FROM users WHERE name = 'Bob'")
conn.commit()

# Verify deletion
cursor.execute("SELECT * FROM users")
print("Remaining Records:")
for row in cursor.fetchall():
    print(row)

# Finally, close the connection when done
conn.close()
```
