# Pandas

Pandas is a fundamental library for data manipulation and analysis in Python. It provides high-performance, easy-to-use data structures like Series and DataFrames. In this lesson, we will cover the basics of Pandas, focusing on data structures, basic operations, and data exploration.

```python
import pandas as pd
import numpy as np

## create a Series using pandas
s = pd.Series([1, 2, 3, 4, 5])
print(s)
print(type(s))
```

```python
## create a DataFrame using pandas
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 27, 22, 32, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

## Basic Data Exploration

```python
## View the first few rows
print(df.head(2))
```

```python
## Get information about the DataFrame
print(df.info())
```

```python
## Get statistical summary of numeric columns
print(df.describe())
```

## DataFrame Slicing and Indexing

```python
## Select a single column
print(df['Age'])
```

```python
## loc: select by label
print(df.loc[0])
```

```python
## iloc: select by index (integer position)
print(df.iloc[1:3])
```

## Logical Operations (Filtering)

```python
## Filter based on a condition
print(df[df['Age'] > 25])
```
